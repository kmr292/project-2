```python
import pandas as pd
```

```python
df = pd.read_csv("netflix_titles(original).csv")
df
```

```python
df.isna().sum()
```

```python
del df['director']
del df["show_id"]
del df["date_added"]
df.head()
```

```python
df= df.dropna()
df
```

```python
df_us= df[df['country'].str.contains("United States")]
df_us
```

```python
df_us = df_us.sort_values('release_year', ascending=False)
df_us
```

```python
df_us = df_us.rename(columns={'title':'TITLE', 'type':'TYPE', 'cast':'CAST', 'country':'COUNTRY PRODUCED',
                              'release_year': 'RELEASE YEAR', 'rating': 'RATING','duration':'DURATION', 'listed_in':'GENRE',
                              'description':'DESCRIPTION'})
df_us
```

```python
df_us = df_us[["TITLE", "TYPE", "RELEASE YEAR", "COUNTRY PRODUCED", "CAST", "RATING", "DURATION", "GENRE", "DESCRIPTION"]]
df_us
```
